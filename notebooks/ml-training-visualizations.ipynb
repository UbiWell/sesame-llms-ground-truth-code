```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
drive_base = "../"
```

#### XGBoost performance comparison between GPT-4 and original responses

```python
fig, axs = plt.subplots(1, 2, figsize=(9, 4))

for j, label in enumerate(["phq9", "gad7"]):
    array1 = [[],[],[],[],[]]
    array2 = [[],[],[],[],[]]

    days = ['5', '6', '7', '8', '9', '10']
    models = ["XG"]
    week = 2

    for split in days:
        df = pd.read_csv(drive_base + "DepreST-CAT-main/machineLearning/screeningResultsGPT4/resultsCAT"+ str(week) + "week" + label + "split" + str(split) + ".csv").fillna(0)
        df2 = pd.read_csv(drive_base + "DepreST-CAT-main/machineLearning/screeningResultsOriginal/resultsCAT"+ str(week) + "week" + label + "split" + str(split) + ".csv").fillna(0)


        for m in range(0, len(models)):
            d = df[df.model == models[m]]
            d1 = d[d.nFeatures == 1]
            array1[m].append(sum(d1.F1)/100)

            d2 = df2[df2.model == models[m]]
            d2_1 = d2[d2.nFeatures == 1]
            array2[m].append(sum(d2_1.F1)/100)


    for i in range(len(models)):
        axs[j].plot(days, array1[i], marker='o', label='GPT-4 scores', linewidth=3, markersize=8)
        axs[j].plot(days, array2[i], marker='X', label='Original scores', linewidth=3, markersize=8)

        axs[j].set_xlabel(label[:-1].upper() + "-" + label[-1] + " Cutoff", fontsize=14)
        if j == 0:
            axs[j].set_ylabel("Average F1", fontsize=14)
        if label == "phq9":
            axs[j].set_title("Depression Screening", fontsize=15)
        else:
            axs[j].set_title("Anxiety Screening", fontsize=15)
        axs[j].set_ylim(0.35, 0.8)

        axs[j].legend(fontsize='large', loc="lower left", frameon=False)
    for tick in axs[j].xaxis.get_major_ticks():
        tick.label1.set_fontsize(14)
    for tick in axs[j].yaxis.get_major_ticks():
        tick.label1.set_fontsize(14)

plt.tight_layout()
plt.show()
plt.close()
```

#### All models performance comparison between GPT-4 and original responses

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

array1 = [[],[],[]]
array2 = [[],[],[]]

days = ['5','6','7','8','9', '10']

models = ["RF", "XG", "kNN", "LR", "SVC",]
legend_labels = ['GPT-4 Simulated Scores', 'Original Scores']

label = "phq9"
week = 2

# Initialize lists to store F1 scores
for _ in range(len(models)):
    array1.append([])
    array2.append([])

for split in days:
    df = pd.read_csv(drive_base + "DepreST-CAT-main/machineLearning/screeningResultsGPT4/resultsCAT"+ str(week) + "week" + label + "split" + str(split) + ".csv").fillna(0)
    df2 = pd.read_csv(drive_base + "DepreST-CAT-main/machineLearning/screeningResultsOriginal/resultsCAT"+ str(week) + "week" + label + "split" + str(split) + ".csv").fillna(0)
    # Calculate F1 scores and store them in respective arrays
    for m in range(len(models)):
        d = df[df.model == models[m]]
        d1 = d[d.nFeatures == 1]
        array1[m].append(sum(d1.F1)/100)

        d2 = df2[df2.model == models[m]]
        d2_1 = d2[d2.nFeatures == 1]
        array2[m].append(sum(d2_1.F1)/100)

plt.figure(figsize=(12, 8))

for i in range(len(models)):
    plt.subplot(2, 3, i+1)
    plt.plot(days, array1[i], marker='o')
    plt.plot(days, array2[i], marker='X')
    plt.xlabel("PHQ-9 Cutoff")
    plt.ylabel("Average F1")
    plt.title(models[i])
    plt.ylim(0.35, 0.85)

plt.figlegend(legend_labels, loc='upper center', ncol=2, bbox_to_anchor=(0.5, 1.05))

plt.suptitle('Depression Screening Models', fontsize=16)

plt.tight_layout()

plt.show()
plt.close()
```
